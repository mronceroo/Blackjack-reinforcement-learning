##### Juego con la máquina

```python
# Recuperamos el mejor agente entrenado
mejor_agente = Maquina(exploracion=0.0)  

# Definimos la ruta donde se guardó la política
mas_entrenamientos = 50000  # Número de entrenamientos realizados
ruta_politica = Path(ARTEFACTOS_PATH) / f"blackjack_politica_{mas_entrenamientos}.yml"

# Instancia del juego de Blackjack
juego_test = JuegoBlackjack(mejor_agente)

# Jugar la partida
juego_test.jugar_con_humano()
```

### Evaluación de la Máquina

#### Porcetaje de victorias

```python
# Evaluamos el desempeño de la máquina
n_partidas = 50000  

porcentaje_victorias = juego_test.evaluacion(episodios=n_partidas)

# Mostramos el resultado
print(f"Evaluación completada: El agente ganó {porcentaje_victorias:.2f}% de las {n_partidas} partidas.")
```

La máquina gana entre el 89% y el 90% de las 50000 partidas

#### Resultados

```python
# Simula varias partidas y registra si la máquina gana o pierde
entrenamientos = 50000
resultados = []  # 1 para victoria, 0 para derrota

for _ in range(entrenamientos):
    juego_test.juego.reset()  # Reinicia el juego
    total_maquina = 0
    
    # Turno de la máquina
    while True:
        accion = juego_test.agente.decidir_accion(total_maquina)  # Acción del agente
        if accion == 1:  # Pedir carta
            carta = juego_test.juego.pedir_carta()
            total_maquina += carta
            if total_maquina > juego_test.juego.objetivo:  # Se pasa
                resultados.append(0)  # Derrota
                break
        else:  # Plantarse
            resultados.append(1)  # Victoria
            break

# Graficamos los resultados como un histograma
plt.hist(resultados, bins=[-0.5, 0.5, 1.5], edgecolor='black', rwidth=0.8)
plt.xticks([0, 1], ['Derrotas', 'Victorias'])
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.title(f'Resultados de {entrenamientos} partidas de Blackjack')
plt.show()
```
